```python
import time
import os
from os.path import exists

def count(dir, counter=0):
    "returns number of files in dir and subdirs"
    for pack in os.walk(dir):
        for f in pack[2]:
            counter += 1
    return dir + " : " + str(counter) + "files"
```

```python
for angry_img in os.listdir(main_path_anger):
  path = main_path_anger + '\\' + angry_img
  img = cv2.imread(path)
  
  if img is None:
    break

  gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  
  faces = detector(gray, 1)
  for face in faces:
      landmarks = predictor(gray, face)
    
  choice1 = color_blue

  points = []
  for i in range(1, 16):
      point = [landmarks.part(i).x, landmarks.part(i).y]
      points.append(point)

  mask_c = [((landmarks.part(29).x), (landmarks.part(29).y))]

  fmask_c = points + mask_c
  fmask_c = np.array(fmask_c, dtype=np.int32)
  img2 = cv2.polylines(img, [fmask_c], True, choice1, thickness=2, lineType=cv2.LINE_8)
  img3 = cv2.fillPoly(img2, [fmask_c], choice1, lineType=cv2.LINE_AA)

  # output
  outputNameofImage = mask_anger + '\\' + angry_img
  print("Saving output image to", outputNameofImage)
  cv2.imwrite(outputNameofImage, img3)
```

```python
for angry_img_crop in os.listdir(main_path_anger):
  path = main_path_anger + '\\' + angry_img_crop
  print(path)
  img = cv2.imread(path)
  assert not isinstance(img,type(None)), 'image not found'
  

  gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  
  faces = detector(gray, 1)
  for face in faces:
      landmarks = predictor(gray, face)
    
  choice1 = color_blue

  points = []
  for i in range(1, 16):
      point = [landmarks.part(i).x, landmarks.part(i).y]
      points.append(point)

  mask_c = [((landmarks.part(29).x), (landmarks.part(29).y))]

  fmask_c = points + mask_c
  fmask_c = np.array(fmask_c, dtype=np.int32)
  img2 = cv2.polylines(img, [fmask_c], True, choice1, thickness=2, lineType=cv2.LINE_8)
  img3 = cv2.fillPoly(img2, [fmask_c], choice1, lineType=cv2.LINE_AA)

  if(landmarks.part(20).y>landmarks.part(25).y):
      pointA1=[((landmarks.part(1).x-20),(landmarks.part(20).y)-50)]
  else:
      pointA1=[((landmarks.part(1).x-20),(landmarks.part(25).y)-50)]
  pointB1=[((landmarks.part(16).x+20),(landmarks.part(31).y-40))]

  cropWindow2=pointA1+pointB1
  cropWindow2=np.array(cropWindow2,dtype=np.int32)

  #print(landmarks.part(20).y)
  #print(landmarks.part(25).y)
  #print(cropWindow2)

  x1 = cropWindow2[0][0] # x1
  y1 = cropWindow2[0][1] # y1

  x2 = cropWindow2[1][0] # x2
  y2 = cropWindow2[1][1] # y2

  #print(x1,y1)
  #print(y1,y2)

  if (x1 < 0 or x2 < 0 or y1 < 0 or y2 < 0):
    if x1 < 0:
      x1 = 0
    elif y1 < 0:
      y1 = 0
    elif x2 < 0:
      x2 = 0
    else:
      y2 = 0

  # crop_img2=img3[cropWindow2[0][1]:cropWindow2[1][1],cropWindow2[0][0]:cropWindow2[1][0]]

  img3 = img3[y1:y2,x1:x2]
  #plt.imshow(crop_img2)

  if img3 is None:
    print("is None")
    break
  
  # output
  outputNameofImage = mask_crop_anger + '\\' + angry_img_crop
  if outputNameofImage is None:
    print("is None")
    break
  print("Saving output image to", outputNameofImage)
  temp=[]
  try:
    cv2.imwrite(outputNameofImage, img3)
  except:
    temp.append(angry_img_crop)
```
