Importing necessary libraries

```python
import cv2
import dlib
import numpy as np
import os
import imutils
```

Setting up the main directory and colour values of the mask

```python
os.chdir('/content/drive/MyDrive/Emotion-Detection-In-Masked-Images')
color_blue = (239,207,137)
```

  Instantiation of Face Detector

```python
detector = dlib.get_frontal_face_detector()
p = '/content/drive/MyDrive/Emotion-Detection-In-Masked-Images/shape_predictor_68_face_landmarks.dat'
predictor = dlib.shape_predictor(p)
```

Original Dataset Path

```python
main_path_anger = '/content/drive/MyDrive/Emotion-Detection-In-Masked-Images/dataset/anger'
main_path_happy = '/content/drive/MyDrive/Emotion-Detection-In-Masked-Images/dataset/happy'
main_path_sadness = '/content/drive/MyDrive/Emotion-Detection-In-Masked-Images/dataset/sadness'
```

Masked Dataset Path

```python
new_path_anger = '/content/drive/MyDrive/Emotion-Detection-In-Masked-Images/dataset/anger_masked'
new_path_happy = '/content/drive/MyDrive/Emotion-Detection-In-Masked-Images/dataset/happy_masked'
new_path_sadness = '/content/drive/MyDrive/Emotion-Detection-In-Masked-Images/dataset/sadness_masked'
```

Masking Anger images

```python
for angry_img in os.listdir(main_path_anger):

  path = main_path_anger + '/' + angry_img
  img = cv2.imread(path)

  if img is None:
    break

  gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  
  faces = detector(gray, 1)
  for face in faces:
      landmarks = predictor(gray, face)
    
  choice1 = color_blue

  points = []
  for i in range(1, 16):
      point = [landmarks.part(i).x, landmarks.part(i).y]
      points.append(point)

  mask_c = [((landmarks.part(29).x), (landmarks.part(29).y))]

  fmask_c = points + mask_c
  fmask_c = np.array(fmask_c, dtype=np.int32)
  img2 = cv2.polylines(img, [fmask_c], True, choice1, thickness=2, lineType=cv2.LINE_8)
  img3 = cv2.fillPoly(img2, [fmask_c], choice1, lineType=cv2.LINE_AA)

  # output
  outputNameofImage = new_path_anger + '/' + angry_img
  print("Saving output image to", outputNameofImage)
  cv2.imwrite(outputNameofImage, img3)
```
